```python
import random as rnd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from termcolor import colored
```

```python
def print_array_info(arra):
    color = 'green'
    print('Array:', colored(arra, color))
    print('Size:', colored(arra.size, color), 'equals length', colored(len(arra), color))
    print('Shape:', colored(arra.shape, color))
    print('Data type:', colored(arra.dtype, color))
    print('Dimention count:', colored(arra.ndim, color))
```
