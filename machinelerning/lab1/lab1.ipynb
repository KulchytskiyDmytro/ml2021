```python
print("///////////////////////////////////////////")
print("Create")
array = np.arange(0, 225, 5)
print("///////////////////////////////////////////")
print("Read")
print()
print(array)
print("///////////////////////////////////////////")
print("Update")
print()
array = np.insert(array, 25, 25)
print(array)
print("///////////////////////////////////////////")
print("Delete")
print()
array = np.delete(array, 0)
print(array)
print("///////////////////////////////////////////")
print("Using random")
print()
secondArray = np.random.randint(25, size=(array.size))
print(secondArray)
print("///////////////////////////////////////////")
```

```python
print("///////////////////////////////////////////")
print("Create")
matrix = np.arange(16).reshape(4, 4)
print("///////////////////////////////////////////")
print("Read")
print()
print(matrix)
print("///////////////////////////////////////////")
print("Update")
print()
newrow = [25,25,25,25]
row_n = matrix.shape[0] ##last row
matrix = np.insert(matrix,row_n,[newrow],axis= 0)
print(matrix)
print("///////////////////////////////////////////")
print("Delete")
print()
matrix = np.delete(matrix, 0, 0)
print(matrix)
print("///////////////////////////////////////////")
```
